$$ |\psi_0(x)|^2 = \frac{1}{Z}\lim_{\tau\to\infty}\int e^{-\varepsilon\mathcal{E}}dx_1\cdots dx_{N-1} $$
$$ Z = \lim_{\tau\to\infty}\int e^{-\varepsilon\mathcal{E}}dx dx_1\cdots dx_{N-1} $$

con $$\mathcal{P} = e^{-\varepsilon\mathcal{E}} = e^{-\mathcal{E} / k_B T}, \quad \Longrightarrow \quad k_B T = \frac{1}{\varepsilon} \equiv \frac{\hbar}{\varepsilon} $$

```python
# parametros de la simulacion

N = 200 ; eps = 0.1
M = 201 ; delta = 1.8*2/(M-1)

# grilla espacial y temporal

time_grid = eps*np.arange(N+1)
space_grid = np.linspace(-1.8, 1.8, M)

# potencial armonico simple

def V(x): return x**2 / 2

def E(x1, x2, eps, V, m=1):
    """
    Retorna la energia en (x1 + x2)/2
    """
    T = 0.5*m*((x1 - x2)/eps)**2    # energia cinetica

    U = V((x1 + x2)/2)              # energia potencial

    return T + U


path = np.zeros(N+1)                # camino inicial (cercano a la solucion) de puros ceros
distribution = np.zeros(M)          # grilla donde se guardan las posiciones aceptadas

for k in range(200):

    oldE = 0 ; newE = 0 ; change = 0 ; trials = 10000

    path = np.zeros(N+1)
    path[-1] = path[0]

    for i in range(trials):

        element = np.random.randint(1, N)
        U = np.random.uniform()

        x_element = path[element].copy()

        oldE = E(path[element-1], path[element], eps, V) + E(path[element], path[element+1], eps, V)

        change = np.random.randint(0, M)

        path[element] = space_grid[change]
        newE = E(path[element-1], path[element], eps, V) + E(path[element], path[element+1], eps, V)

        if newE >  oldE and np.exp(-eps*(newE - oldE)) < U:     # rechazamos
            path[element] = x_element
        # else: oldE = newE

        element = np.where(space_grid == path[element])[0]
        if element < M and element >= 0: distribution[element] += 1

    
```

## Implementación 2D de integrales de camino para obtener el primer auto-estado

```python
# parametros de la simulacion
# grilla espacial y temporal
# potencial armonico simple

# parametros de la simulacion

N = 40 ; eps = 0.1
M = 41 ; delta = 1.8*2/(M-1)

# grilla espacial y temporal

time_grid = eps*np.arange(N+1)
x_space_grid = np.linspace(-1.8, 1.8, M)
y_space_grid = np.linspace(-1.8, 1.8, M)

# potencial armonico simple 2D

def V(x, y): return (x**2 + y**2) / 2

def E(x1, x2, eps, V, m=1):
    """
    x1:     Posiciones [x,y] de particula en t1 (array)
    x2:     Posiciones [x,y] de particula en t2 (array)
    eps:    espaciado temporal (float)
    V:      Potencial (function)
    m:      masa particula (float)
    ---
    return:
    E:      Energia en el punto medio
    """
    T = 0.5*m*((x1[0] - x2[0])/eps)**2 + 0.5*m*((x1[1] - x2[1])/eps)**2     # energia cinetica

    U = V((x1[0] + x2[0])/2, (x1[0] + x2[0])/2)         # energia potencial

    return T + U



path = np.zeros((N+1, 2))               # camino inicial (cercano a la solucion) de ceros
distribution = np.zeros((M, M))         # grilla donde se guardan las posiciones aceptadas

#for k in range(200):

oldE = 0 ; newE = 0 ; change = 0 ; trials = 4000000

path = np.zeros((N+1, 2))
path[-1] = path[0]

for i in range(trials):

    # seleccionamos un tiempo aleatorio (excluyendo el inicio y final)
    t_element = np.random.randint(1, N)

    U = np.random.uniform()

    # calculamos la energia del camino en ese tiempo
    oldE = E(path[t_element-1], path[t_element], eps, V) + E(path[t_element], path[t_element+1], eps, V)

    x_element = path[t_element][0].copy()
    y_element = path[t_element][1].copy()

    # seleccionamos posiciones aleatorias espaciales
    x_change = np.random.randint(0, M)
    y_change = np.random.randint(0, M)

    # cambiamos valores del camino en tiempo t_element
    path[t_element] = x_space_grid[x_change], y_space_grid[y_change]
    
    # calculamos nueva energia en t_element
    newE = E(path[t_element-1], path[t_element], eps, V) + E(path[t_element], path[t_element+1], eps, V)

    # aplicamos Hasting
    if newE >  oldE and np.exp(-eps*(newE - oldE)) < U:     # rechazamos
        path[t_element] = x_element, y_element
    #else: oldE = newE

    x_element = np.where(x_space_grid == path[t_element][0])[0]
    y_element = np.where(y_space_grid == path[t_element][1])[0]

    # guardamos el que fue escogido
    if max(x_element,y_element) < M and min(x_element, y_element) >= 0:
        
        distribution[x_element, y_element] += 1

```

```python
z = distribution / np.sum(distribution)
x, y = z.nonzero()

fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z[z != 0], cmap='viridis', edgecolor='none')
plt.show()

```

```python
# parametros de la simulacion

N = 30 ; eps = 0.1
M = 31 ; delta = 1.8*2/(M-1)

# grilla espacial y temporal

time_grid = eps*np.arange(N+1)
x_space_grid = np.linspace(-1.8, 1.8, M)
y_space_grid = np.linspace(-1.8, 1.8, M)

# potencial de repulsive short-range and attractive mid-range nucleon-nucleon

def V(x, y):
    
    V1 = 12 ; V2 = -12 ; sigma1 = 0.2 ; sigma2 = 0.8

    A = V1*np.exp(-(x**2 + y**2) / (sigma1**2)) / (sigma1*np.sqrt(np.pi))
    B = V2*np.exp(-(x**2 + y**2) / (sigma2**2)) / (sigma2*np.sqrt(np.pi))

    return A + B

def E(x1, x2, eps, V, m=1):
    """
    x1:     Posiciones [x,y] de particula en t1 (array)
    x2:     Posiciones [x,y] de particula en t2 (array)
    eps:    espaciado temporal (float)
    V:      Potencial (function)
    m:      masa particula (float)
    ---
    return:
    E:      Energia en el punto medio
    """
    T = 0.5*m*((x1[0] - x2[0])/eps)**2 + 0.5*m*((x1[1] - x2[1])/eps)**2

    U = V((x1[0] + x2[0])/2, (x1[1] + x2[1])/2)

    return T + U



path = np.zeros((N+1, 2))               # camino inicial (cercano a la solucion) de ceros
distribution = np.zeros((M, M))         # grilla donde se guardan las posiciones aceptadas

#for k in range(200):

oldE = 0 ; newE = 0 ; change = 0 ; trials = 4000000

path = np.zeros((N+1, 2))
path[-1] = path[0]

for i in range(trials):

    # seleccionamos un tiempo aleatorio (excluyendo el inicio y final)
    t_element = np.random.randint(1, N)

    U = np.random.uniform()

    # calculamos la energia del camino en ese tiempo
    oldE = E(path[t_element-1], path[t_element], eps, V) + E(path[t_element], path[t_element+1], eps, V)

    x_element = path[t_element][0].copy()
    y_element = path[t_element][1].copy()

    # seleccionamos posiciones aleatorias espaciales
    x_change = np.random.randint(0, M)
    y_change = np.random.randint(0, M)

    # cambiamos valores del camino en tiempo t_element
    path[t_element] = x_space_grid[x_change], y_space_grid[y_change]
    
    # calculamos nueva energia en t_element
    newE = E(path[t_element-1], path[t_element], eps, V) + E(path[t_element], path[t_element+1], eps, V)

    # aplicamos Hasting
    if newE >  oldE and np.exp(-eps*(newE - oldE)) < U:     # rechazamos
        path[t_element] = x_element, y_element
    #else: oldE = newE

    x_element = np.where(x_space_grid == path[t_element][0])[0]
    y_element = np.where(y_space_grid == path[t_element][1])[0]

    # guardamos el que fue escogido
    if max(x_element,y_element) < M and min(x_element, y_element) >= 0:
        
        distribution[x_element, y_element] += 1
```

```python
# parametros de la simulacion

N = 30 ; eps = 0.1
M = 31 ; delta = 1.8*2/(M-1)

# grilla espacial y temporal

time_grid = eps*np.arange(N+1)
x_space_grid = np.linspace(-1.8, 1.8, M)
y_space_grid = np.linspace(-1.8, 1.8, M)

# potencial de repulsive short-range and attractive mid-range nucleon-nucleon

def V(x, y):

    return ()
    
    # V1 = 12 ; V2 = -12 ; sigma1 = 0.2 ; sigma2 = 0.8

    # A = V1*np.exp(-(x**2 + y**2) / (sigma1**2)) / (sigma1*np.sqrt(np.pi))
    # B = V2*np.exp(-(x**2 + y**2) / (sigma2**2)) / (sigma2*np.sqrt(np.pi))


def E(p1x1, p1x2, p2x1, p2x2, eps, V, m=1):
    """
    p1x1:     Posiciones [x,y] de particula 1 en t1 (array)
    p1x2:     Posiciones [x,y] de particula 1 en t2 (array)
    p2x1:     Posiciones [x,y] de particula 2 en t1 (array)
    p2x2:     Posiciones [x,y] de particula 2 en t2 (array)
    eps:    espaciado temporal (float)
    V:      Potencial (function)
    m:      masa particula (float)
    ---
    return:
    E:      Energia en el punto medio
    """
    T_p1 = ((p1x1[0] - p1x2[0])/eps)**2 + ((p1x1[1] - p1x2[1])/eps)**2
    T_p2 = ((p2x1[0] - p2x2[0])/eps)**2 + ((p2x1[1] - p2x2[1])/eps)**2

    T = 0.5*m*(T_p1 + T_p2)

    p1xm = (p1x1[0] + p1x2[0])/2 ; p1ym = (p1x1[1] + p1x2[1])/2
    p2xm = (p2x1[0] + p2x2[0])/2 ; p2ym = (p2x1[1] + p2x2[1])/2

    U = V(p1xm - p2xm, p1ym - p2ym)

    return T + U



path = np.zeros((2, N+1, 2))            # 2 caminos iniciales (cercanos a la solucion) de ceros
distribution = np.zeros((2, M, M))      # 2 grillas donde se guardan las posiciones aceptadas

#for k in range(200):

oldE = 0 ; newE = 0 ; change = 0 ; trials = 4000000

path = np.zeros((2, N+1, 2))

for i in range(trials):

    # seleccionamos un tiempo aleatorio (excluyendo el inicio y final)
    t_element = np.random.randint(1, N)

    # seleccionamos un camino aleatorio
    index_path = np.random.randint(0, 2)

    U = np.random.uniform()

    # calculamos la energia del sistema en ese tiempo
    oldE = E(path[0][t_element-1], path[0][t_element], path[1][t_element-1], path[1][t_element], eps, V) + E(path[0][t_element], path[0][t_element+1], path[1][t_element], path[1][t_element+1], eps, V)

    x_element = path[index_path][t_element][0].copy()
    y_element = path[index_path][t_element][1].copy()

    # seleccionamos posiciones aleatorias espaciales
    x_change = np.random.randint(0, M)
    y_change = np.random.randint(0, M)

    # cambiamos valores del camino en tiempo t_element
    path[index_path][t_element] = x_space_grid[x_change], y_space_grid[y_change]
    
    # calculamos nueva energia en t_element del sistema de 2 particulas
    newE = E(path[0][t_element-1], path[0][t_element], path[1][t_element-1], path[1][t_element], eps, V) + E(path[0][t_element], path[0][t_element+1], path[1][t_element], path[1][t_element+1], eps, V)

    # aplicamos Hasting
    if newE >  oldE and np.exp(-eps*(newE - oldE)) < U:     # rechazamos
        path[index_path][t_element] = x_element, y_element
    #else: oldE = newE

    x_element = np.where(x_space_grid == path[index_path][t_element][0])[0]
    y_element = np.where(y_space_grid == path[index_path][t_element][1])[0]

    # guardamos el que fue escogido
    if max(x_element,y_element) < M and min(x_element, y_element) >= 0:
        
        distribution[index_path,x_element, y_element] += 1
```

```python
N = 40 ; eps = 0.1
M = 41 ; delta = 1.8*2/(M-1)

time_grid = eps*np.arange(N+1)
x_space_grid = np.linspace(-1.8, 1.8, M)

def V(x): return (x**2) / 2

def E(p1x1, p1x2, p2x1, p2x2, eps, V, m=1):
    """
    p1x1:     Posiciones [x] de particula 1 en t1 (array)
    p1x2:     Posiciones [x] de particula 1 en t2 (array)
    p2x1:     Posiciones [x] de particula 2 en t1 (array)
    p2x2:     Posiciones [x] de particula 2 en t2 (array)
    eps:    espaciado temporal (float)
    V:      Potencial (function)
    m:      masa particula (float)
    ---
    return:
    E:      Energia en el punto medio
    """
    T_p1 = ((p1x1 - p1x2)/eps)**2
    T_p2 = ((p2x1 - p2x2)/eps)**2

    T = 0.5*m*(T_p1 + T_p2)

    p1m = (p1x1 + p1x2)/2
    p2m = (p2x1 + p2x2)/2

    U = V(p1m - p2m)

    return T + U


path = np.zeros((2, N+1))            # 2 caminos iniciales (cercanos a la solucion) de ceros
distribution = np.zeros((2, M))      # 2 grillas donde se guardan las posiciones aceptadas

#for k in range(200):

oldE = 0 ; newE = 0 ; change = 0 ; trials = 4000000

path = np.zeros((2, N+1))

for i in range(trials):

    # seleccionamos un tiempo aleatorio (excluyendo el inicio y final)
    t_element = np.random.randint(1, N)

    # seleccionamos un camino aleatorio
    index_path = np.random.randint(0, 2)

    U = np.random.uniform()

    # calculamos la energia del sistema en ese tiempo
    oldE = E(path[0][t_element-1], path[0][t_element], path[1][t_element-1], path[1][t_element], eps, V) + E(path[0][t_element], path[0][t_element+1], path[1][t_element], path[1][t_element+1], eps, V)

    x_element = path[index_path][t_element].copy()

    # seleccionamos posiciones aleatorias espaciales
    x_change = np.random.randint(0, M)

    # cambiamos valores del camino en tiempo t_element
    path[index_path][t_element] = x_space_grid[x_change]
    
    # calculamos nueva energia en t_element del sistema de 2 particulas
    newE = E(path[0][t_element-1], path[0][t_element], path[1][t_element-1], path[1][t_element], eps, V) + E(path[0][t_element], path[0][t_element+1], path[1][t_element], path[1][t_element+1], eps, V)

    # aplicamos Hasting
    if newE >  oldE and np.exp(-eps*(newE - oldE)) < U:     # rechazamos
        path[index_path][t_element] = x_element
    #else: oldE = newE

    x_element = np.where(x_space_grid == path[index_path][t_element])[0]

    # guardamos el que fue escogido
    if x_element < M and x_element >= 0:
        
        distribution[index_path, x_element] += 1
```

```python
def update(x, eps, L):
    """
    Realiza un paso de MC Hasting
    """
    for j in range(0,N):
        old_x = x[j]
        old_Sj = S(j, x, eps)
        x[j] += np.random.uniform(-L, L)
        deltaS = S(j, x, eps) - old_Sj

        if deltaS > 0 and np.exp(-deltaS) < np.random.uniform():    # reject
            x[j] = old_x
        
def S(j, x, eps):
    jn = (j+1)%N            # next index
    jp = (j-1)%N            # previous index
    U = eps*x[j]**2 / 2
    T = x[j]*(x[j] - x[jn] - x[jp])/eps
    return T + U


```

Tomando $$G(t) \equiv \langle \langle x(t_2)x(t_1)\rangle\rangle \equiv \frac{\int \mathcal{D}x(t)x(t_2)x(t_1)e^{-S[x]}}{\int \mathcal{D}x(t)e^{-S[x]}} = \int \langle x|e^{-\tilde{H} (t_f - t_2)}\tilde{x} e^{-\tilde{H} (t_2 - t_1)} \tilde{x} e^{-\tilde{H} (t_1 - t_i)}|x\rangle dx$$ con $t = t_2 - t_1$, $T = t_f - t_i$.

$$ \langle \langle x(t_2)x(t_1)\rangle\rangle = \frac{\sum e^{-E_n T}\langle E_n | \tilde x e^{-(\tilde H - E_n)t}\tilde x | E_n\rangle}{\sum e^{-E_n T}}$$

Sabemos que si hacemos $T, t$ grandes pero con $t\ll T$, tenemos que

$$ G(t) \longrightarrow \langle E_0 | \tilde x e^{-(\tilde H - E_0)t}\tilde x | E_0\rangle \approx |\langle E_0 | \tilde x | E_1 \rangle|^2 e^{-(E_1 - E_0)t}.$$

Con lo cual tenemos $$ \log(G(t) / G(t+\varepsilon)) \longrightarrow (E_1 - E_0)\varepsilon $$

Consideraremos en nuestro caso $$G(t) = \frac{1}{N}\sum_j \langle\langle x(t_j + t) x_j\rangle\rangle  $$ para $t = 0, \varepsilon, 2\varepsilon, ...., (N-1)\varepsilon$. Es decir,

$$ G_n = \frac{1}{N}\sum_j \langle\langle x_{(t_j + n)\mod{N}}\ x_j\rangle\rangle $$ para $n=0,...,N-1$.

```python
# parametros de la simulacion

eps = 0.5 ; N = 20 ; L = 1.4 ; N_cor = 20

N_cf = [25, 100, 1000, 10000]

def compute_G(x, n, N=20):
    """
    Obtiene el promedio del propagador G = x(t2)*x(t1)
    """
    G = 0
    for j in range(N):
        G += x[(j+n)%N]*x[j]

    return G/N


def MCaverage(x, comp_G, N_cf, N=20):
    G = np.zeros((N_cf, N))
    avg_G = np.zeros(N)

    for j in range(N):
        x[j] = 0

    for j in range(0, 5*N_cor):
        update(x, eps, L)

    for alpha in range(N_cf):
        for j in range(N_cor):
            update(x, eps, L)
        for n in range(N):
            G[alpha, n] = comp_G(x, n)

    for n in range(N):
        avg_G[n] = G[:, n].mean()
    
    return G, avg_G

```

```python
n_bin = 10

def binning(G):
    bin_G = []

    for i in range(0, len(G), n_bin):
        sum_bin_G = 0
        for j in range(n_bin):
            sum_bin_G += G[i + j]
        bin_G.append(sum_bin_G / n_bin)

    return np.asarray(bin_G)

```

```python
n_bootstraps = 100

def bootstrap(G):
    G_bootstraps = G.copy()

    for i in range(len(G)):
        G_bootstraps[i] = G[np.random.randint(0, len(G))]

    return G_bootstraps

def avg_G(GG):
    avgG = np.zeros(N) 
    for n in range(N):
        avgG[n] = GG[:, n].mean()
    
    return avgG

def errores(avg_G, avg_G_bt):

    err = np.zeros(N-1)
    dE = np.zeros(N-1)

    for i in range(N-1):
        dEi_bootstraps = np.zeros(n_bootstraps)

        dE[i] = np.log(avg_G[i] / avg_G[i+1]) / eps

        for j in range(n_bootstraps):
            dEi_bootstraps[j] = np.log(avg_G_bt[j][i] / avg_G_bt[j][i+1]) / eps

        err[i] = np.std(dEi_bootstraps)
    
    return dE, err


eps = 0.5 ; N = 20 ; L = 1.4 ; N_cor = 20

N_cf = [25, 100, 1000, 10000]

def compute_G(x, n):
    G = 0
    for j in range(N):
        G += x[(j+n)%N]*x[j]

    return G/N

G = np.zeros((N_cf[3], N))


def MCaverage(x, G, comp_G):
    avgG = np.zeros(N)
    for j in range(N):
        x[j] = 0

    for j in range(0, 5*N_cor):
        update(x, eps, L)

    for alpha in range(len(G)):
        for j in range(N_cor):
            update(x, eps, L)
        for n in range(N):
            G[alpha, n] = comp_G(x, n)

    for n in range(N):
        avgG[n] = G[:, n].mean()

    return avgG
```

```python
eps = 0.5 ; N = 20 ; L = 1.4 ; N_cor = 20

N_cf = [25, 100, 1000, 10000]
N_bin = [1, 5, 10, 100]
N_bootstraps = [100, 100, 100, 100]

dE_lista = []
err_lista = []

def compute_G(x, n):
    G = 0
    for j in range(N):
        G += (x[(j+n)%N]*x[j])**3

    return G/N

for i in range(len(N_cf)):
    n_bin = N_bin[i]
    n_bootstraps = N_bootstraps[i]

    G = np.zeros((N_cf[i], N))

    avgG = MCaverage(np.zeros(N), G, compute_G)

    avg_total_G = []
    avg_G_bt = []

    binning_G = binning(G)

    # llenamos la lista de promedios de bootstraps de G
    for _ in range(n_bootstraps):
        avg_G_bt.append(avg_G(bootstrap(binning_G)))

    avg_total_G = avg_G(G)

    # calculamos los errores y la energía de exitación

    dE, err = errores(avg_total_G, avg_G_bt)

    dE_lista.append(dE)
    err_lista.append(err)

plt.figure(figsize=(10,4))
for i in range(len(N_cf)):
    plt.errorbar(eps*np.arange(N-1), dE_lista[i], yerr=err_lista[i], marker='.', label='$N_{cf}$')

plt.axhline(1, color='r', linestyle='--', label='valor teorico')
plt.legend()
plt.xlim(0, 4)
plt.ylabel('$\\Delta E$')
plt.xlabel('t')
plt.title('Simulacion de energia de exitacion')
plt.show()
```
